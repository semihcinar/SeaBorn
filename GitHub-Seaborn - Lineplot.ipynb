# SeaBorn - Line PLOT

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import scipy
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
print(sns.get_dataset_names())
```

```python
flights = sns.load_dataset("flights")
flights
```

```python
plt.figure(figsize = (12,6))
sns.lineplot(data = flights, x = "year", y = "passengers")
plt.show()
```

```python
only_may = flights.query("month == 'May'")
sns.lineplot(data=only_may, x="year", y="passengers")
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(y=flights.passengers.values, x= flights.index)
```

```python
sns.lineplot(data = flights, x = 'year', y = 'passengers', hue='month')
plt.legend(loc=(1.1,.0))
```

```python
flights_wide = flights.pivot("year", "month", "passengers")
flights_wide.head()
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=flights_wide.T)
```

```python
plt.figure(figsize = (12,6))
sns.lineplot(data = flights, x = 'month', y = 'passengers', hue='year')
plt.legend(loc=(1.1,.0))
plt.show()
```
